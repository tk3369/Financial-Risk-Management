```python
import math
import numpy as np
import frmbook_funcs
```

## Virtual Reality

We will simulate 91.5 years of historical reutrn data with 18.49% annual standard devication:

$\quad VR \sim \mathcal{N}(\mu,\,\sigma^{2})$

$\quad\begin{align*}
\mu = & \; 0 \\
\sigma^2 = & \; 18.49
\end{align*}$

```python
annual_sd = 18.49
target_sd = annual_sd / math.sqrt(12)
target_mean = 0.0

# Number of months of virtual reality
num_months = 91 * 12 + 6
end_period = 2018
beginning_period = end_period - num_months / 12
print("From {0:.2f} to {1:.2f}".format(beginning_period, end_period))

np.random.seed(314159265)
virtual_reality = target_sd * np.random.randn(num_months) + target_mean
print("Target SD = {0:.7f}\nSimulation SD   = {1:.7f}\nSimulation Mean = {2:.7f}".format(
      target_sd, np.std(virtual_reality), np.mean(virtual_reality)))
```

## Generate Date Array

```python
# create equal size marks for the dates
dates = np.linspace(beginning_period, end_period, num=num_months)
```

## Plot Figure

```python
lookbacks = [12, 36, 60]
sample_sd = frmbook_funcs.GenSampleSd(virtual_reality, lookbacks)
stub_offset = 6
colors = ['y-','b-','r-']
frmbook_funcs.PlotSampleSd('Figure 1', dates, sample_sd, stub_offset,\
                           lookbacks, colors)
```
