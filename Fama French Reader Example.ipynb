```python
import pandas as pd
```

```python
url = "http://mba.tuck.dartmouth.edu/pages/faculty/ken.french/ftp/F-F_Research_Data_Factors_CSV.zip"
names_monthly = ["yearmon", "mkt_minus_rf", "SMB", "HML", "RF"]
df_monthly = pd.read_csv(url, skiprows=4, nrows=1098, names=names_monthly)
df_monthly["year"]  = df_monthly["yearmon"] // 100
df_monthly["month"] = df_monthly["yearmon"] % 100
df_monthly["date"]  = df_monthly["year"] + df_monthly["month"] / 12
df_monthly
```
